### Tensor math

```python
x = torch.tensor([1,3,4])
y = torch.tensor([6,7,8])
```

```python
# addition

z1 = torch.empty(3)
torch.add(x, y, out=z1)
z1
```

```python
z2 = torch.add(x, y)
z2
```

```python
z3 = x + y
z3
```

```python
# substraction

z = x - y
z
```

```python
# division

z = torch.true_divide(x, y)
z
```

```python
# inplace operations

t = torch.zeros(3)
t.add_(x)         # t is modified
t
```

```python
t += x  # other way
t
```

```python
# exponent

z = x.pow(2)
z
```

```python
x ** 2    # other way
```

```python
# simple comparison
x < 0
```

```python
x > 3
```

```python
# matrix multiplication

x1 = torch.rand((2,5))
x2 = torch.rand((5,3))

y = torch.mm(x1, x2)
y
```

```python
y2 = x1.mm(x2)
y2
```

```python
# matrix exponentiation

matrix_exp = torch.rand(5, 5)
matrix_exp.matrix_power(3)
```

```python
# element wise multiplication

x = torch.tensor([1,3,4])
y = torch.tensor([6,7,8])

z = x * y
z
```

```python
# dot product

z = torch.dot(x, y)
z
```

```python
# batch multiplication

batch = 32
n = 10
m = 20
p = 30

tensor1 = torch.rand((batch, n, m))
tensor2 = torch.rand((batch, m, p))

output_bmm = torch.bmm(tensor1, tensor2)   # the output shape will be (batch, n, p)
output_bmm.shape
```

### Broadcasting

```python
x1 = torch.rand((5, 5))
x2 = torch.rand((1, 5))
print(x1, x2, sep="\n", end="\n\n\n")

z = x1 - x2
z                   #  the second vector is broadcasted to higher dimention
                    #  just like in numpy
```

### Indexing

```python
batch_size = 10
features = 25
x = torch.rand((batch_size, features))
x[0].shape
```

```python
x[:, 0].shape
```

```python
x[2, 0:10]
```

```python
# we can assign like that also

x[0,0] = 100
x[0][0]
```

```python
# a bit fancy indexing

x = torch.arange(10)
indices = [2,5,8]
x[indices]
```

```python
# in a 2d manner

x = torch.rand((3,4))
rows = torch.tensor([1,0])
cols = torch.tensor([2,3])
x[rows, cols].shape
```

```python
# A bit advanced indexing

x  = torch.arange(10)
x[(x<2) | (x>8)]
```

```python
x[x.remainder(2) == 0]
```
